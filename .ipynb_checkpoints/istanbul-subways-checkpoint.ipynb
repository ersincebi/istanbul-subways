```python
# required libraries for project
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clustering
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    max_ite = 10
    while True or i<max_ite:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)

        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                       for i in range(n_clusters)])

        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        
        centers = new_centers

        return centers, labels
```

```python
def clusters(datasets, column):
    kmns = pd.DataFrame(datasets, columns=['population',column])
    kmeans = KMeans(n_clusters=5).fit(kmns)
    return kmns, kmeans
```

```python
class models:
    def __init__(self,mode,n_neighbors=5):
        self.model_modes = {'guassian':GaussianNB()
                       ,'multy': MultinomialNB()
                       ,'knn':KNeighborsClassifier(n_neighbors=n_neighbors)}
        
        self.model = self.model_modes[mode]
        
    def pipeLine(self):
        make_pipeline(TfidfVectorizer(), self.model)

    def model_predict(self, test):
        self.model.predict(test)

    def model_fit(self, X_train, y_train):
        self.model.fit(X_train, y_train)

    def class_report(self, X_test, y_test):
        classification_report(y_test, self.model_predict(X_test))

    def conf_matrix(self, X_test, y_test):
        confusion_matrix(y_test, self.model_predict(X_test))
```
